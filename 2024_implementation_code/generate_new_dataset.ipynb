```python
def add_data_to_dataset(data, label, x_dataset, y_dataset):
    x_dataset.resize(x_dataset.shape[0] + 1, axis=0)
    x_dataset[-1:] = data
    y_dataset.resize(y_dataset.shape[0] + 1, axis=0)
    y_dataset[-1:] = label
```

```python
x_datas = []
y_datas = []

# for song_list, target in [[test_songs, 'test']]:
# for song_list, target in [[train_songs, 'train']]:
for song_list, target in [[FMA_songs, 'test']]:
# for song_list, target in [[RWC_train, 'test']]:
# for song_list, target in [[RWC_test, 'test']]:
        for song in song_list:
            print(song)
            annotations = []
            for d in labs[song]:
                start, end, cat = d.split(' ')
                start = round(float(start), 4)
                end = round(float(end), 4)
                annotations.append([start, end, cat])

            # song_path = (os.path.join(DATA_DIR, f'./JamendoSource/{target}/{song}'))
            song_path = (os.path.join(DATA_DIR, f'./FMASource/{song}'))
            # song_path = (os.path.join(DATA_DIR, f'./RWCSource/{song}'))
            y, sr = librosa.load(song_path, sr=16000)
            duration = librosa.get_duration(y, sr)
            print(duration)
            for d in range(0, int(duration) * 1000, 1000):
                start_time = d
                end = d + 2000
                if(end <= int(duration) * 1000):
                    print(song, start_time / 1000, end / 1000)
                    audio_dst = y[int(start_time / 1000)*sr:int(end / 1000)*sr]
                    audio_dst = np.reshape(audio_dst, (32000, 1))
                    x_datas.append(audio_dst)

                    for index, annotation in enumerate(annotations):
                        y_data = []
                        if (
                            end >= annotations[index][0] * 1000
                            and end <= annotations[index][1] * 1000
                        ):
                            if annotations[index][2] == 'vocal':
                                for i in range(10):
                                    y_data.append(1.)
                            else:
                                for i in range(10):
                                    y_data.append(0.)
                            break
                        elif (start_time + 500 <= annotations[index][1] * 1000):
                            gapCnt = 0
                            for e in range(500, 1500, 100):
                                st = start_time + e
                                et = st + 100
                                isGap = False
                                if(st <= annotations[index][1] * 1000 
                                   and et <= annotations[index][1] * 1000):
                                    if annotations[index][2] == 'vocal':
                                        y_data.append(1.)
                                    else:
                                        y_data.append(0.)
                                elif(st <= annotations[index][1] * 1000 
                                   and et > annotations[index][1] * 1000):
                                    if (annotations[index][1] * 1000 - st) > (et - annotations[index][1] * 1000):
                                        if annotations[index][2] == 'vocal':
                                            y_data.append(1.)
                                        else:
                                            y_data.append(0.)
                                    else:
                                        if annotations[index][2] == 'vocal':
                                            y_data.append(0.)
                                        else:
                                            y_data.append(1.)
                                else:
                                    if (annotations[index+1][1]*1000 - annotations[index+1][0]*1000) <= 1000:
                                        if (st >= annotations[index+1][0]*1000
                                            and st <= annotations[index+1][1]*1000):
                                            if annotations[index+1][2] == 'vocal':
                                                y_data.append(1.)
                                            else:
                                                y_data.append(0.)
                                            gapCnt += 1
                                            isGap = True
                                        else:
                                            isGap = False
                                    if (isGap == False):
                                        if (gapCnt == 0):
                                            if annotations[index][2] == 'vocal':
                                                y_data.append(0.)
                                            else:
                                                y_data.append(1.)
                                        else:
                                            if annotations[index][2] == 'vocal':
                                                y_data.append(1.)
                                            else:
                                                y_data.append(0.)
                            break
                    y_data = np.array(y_data)
                    y_datas.append(y_data)
```
